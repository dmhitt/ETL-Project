```python
#term = 'Mexican restaurant bar'
#location = 'Astoria, NY'
term = 'restaurant reviews'
location = 'Huntington Beach, CA'
search_limit = 50
response = yelp_api.search_query(term = term,
                                 location = location,
                                 limit = search_limit)
                                 
type(response)
```
