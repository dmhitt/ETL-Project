```python
first_time = True

for index, row in data.iterrows():
    id_business =  row['id']
    response = yelp_api.reviews_query(id = id_business)
    if first_time:
        cols = list(response['reviews'][0].keys())
        cols.append('id_business')
        print(cols)
        data_review = pd.DataFrame(columns=cols)
        first_time = False

    #data_review['id_business'] = id_business
    for biz in response['reviews']:
        print (f"biz = {biz}")
        my_row = {"id":id_business, 
                "url":biz['url'],
                "text":biz['text'],
                "rating":biz['rating'],
                "time_created":biz['time_created'],
                "user":biz['user'],
                "id_business":biz['id']}                    
        #my_row.append(id_business)
        print (f"my_row = {my_row}")
        print ('----')
        data_review=data_review.append(my_row, ignore_index=True)
        my_row.clear()
        
     
    print ("------")
    my_row.clear()
 
```
