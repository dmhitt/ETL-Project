```python
# Dependencies

import requests
import json
import pandas as pd


from config import api_key



from yelpapi import YelpAPI
yelp_api = YelpAPI(api_key)

```

```python
#term = 'Asian restaurant bar'
#location = 'Astoria, NY'
term = 'restaurant'
location = 'Irvine, CA'
search_limit = 50
response = yelp_api.search_query(term = term,
                                 location = location,
                                 limit = search_limit)
type(response)
```
